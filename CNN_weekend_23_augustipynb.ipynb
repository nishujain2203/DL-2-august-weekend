```python
x_train_scaled = x_train/255
x_test_scaled  = x_test/255
```

```python
from  keras.utils import  to_categorical
```

```python
y_train_cat  =   to_categorical(y_train)
y_test_cat  =  to_categorical(y_test)
```

# model  fitting

```python
from  keras.models import Sequential
from  keras.layers import  Convolution2D ,  MaxPool2D ,  Dense  , BatchNormalization ,  Flatten , Dropout
```

```python
nn  =  Sequential()

# add CNN  layers
nn.add(Convolution2D(filters= 16  , kernel_size  = (3 , 3) , input_shape = (28 , 28  ,  1) ))
nn.add(MaxPool2D(pool_size=(2 , 2)))
nn.add(BatchNormalization())
nn.add(Dropout(0.1))


# add CNN  layers
nn.add(Convolution2D(filters= 16  , kernel_size  = (3 , 3) ))
nn.add(MaxPool2D(pool_size=(2 , 2)))
nn.add(BatchNormalization())
nn.add(Dropout(0.1))



# add flatten layer
nn.add(Flatten())


# add hiden  layers
nn.add(Dense(128 ,  activation= 'relu' ,   kernel_initializer =  'he_uniform'  ))
nn.add(Dropout(0.1))

# add hiden  layers
nn.add(Dense(128 ,  activation= 'relu' ,   kernel_initializer =  'he_uniform'  ))
nn.add(Dropout(0.1))



# add output  layer
nn.add(Dense(10  ,  activation= 'softmax'))
```

```python
nn.summary()
```

# compile  

```python
nn.compile(optimizer  = 'adam' ,  loss  =  'categorical_crossentropy' ,  metrics  =  ['accuracy'])
```

# model  fitting

```python
from  keras.callbacks import  EarlyStopping
Er  = EarlyStopping(monitor  = 'val_loss' ,  patience  =  5  ,  verbose  = 1 ,restore_best_weights =  True  )
```

```python
hist  =  nn.fit(x_train_scaled  ,  y_train_cat ,  batch_size  =  128  ,  epochs  =  100 ,
       validation_split= 0.2  ,  callbacks= Er   )
```

```python
plt.plot(hist.history['loss'] ,  c = 'green')
plt.plot(hist.history['val_loss'] ,  c = 'red')
```

# training evaluation
